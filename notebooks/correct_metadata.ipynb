```python
from joblib import Parallel, delayed
import os
import json
import sys
from eegdash import EEGDashDataset

metadata = ['subject', 'session', 'run', 'task', 'age', 'gender', 'sex', 'p_factor', 'externalizing', 'internalizing', 'attention']
count = 0

def process_file(path):
    try:
        with open(path) as f:
            data = json.load(f)
            # extract info
            subject = data['subject']
            task = data['task']
            ds = EEGDashDataset({'subject': subject, 'task': task}, description_fields=metadata)
            record = ds.description.iloc[0].to_dict()
            # resave the description.json file using record
            with open(path, 'w') as f:
                print(f"Saving {path}")
                json.dump(record, f, indent=4)
            return 1
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON in {path}: {e}")
        sys.exit(1)
    return 0

# Collect all description.json file paths
file_paths = []
for root, dirs, files in os.walk('data-no_cz-robust_scaled'):
    for file in files:
        if file == 'description.json':
            file_paths.append(os.path.join(root, file))

# Process files in parallel
count = sum(Parallel(n_jobs=-1)(delayed(process_file)(path) for path in file_paths))

print(f"Processed {count} files.")
```
